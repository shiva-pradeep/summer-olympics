# 1. In how many cities Summer Olympics is held so far?


```python
number_of_cities=len(df['City'].unique())
print("number of cities Summer Olympics is held so far is:",number_of_cities)

cities=[]
for i in df['City'].unique():
    cities.append(i)
cities = pd.DataFrame(cities,columns = ['cities'])
cities
```

# 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold_medals = df[df['Medal'] == 'Gold']
gold_medals['Sport']
no_of_gold_medals=[]
for i in gold_medals['Sport'].unique():
    no_of_gold_medals.append([i,len(df[df['Sport']  == i])])
gold_medals = pd.DataFrame(no_of_gold_medals,columns = ['Sport','No of Gold Medals'])
gold_medals.sort_values(by = 'No of Gold Medals', ascending=False).head()
```

```python
gold_medals.sort_values(by = 'No of Gold Medals', ascending=False).head().plot(x='Sport' , y='No of Gold Medals', kind = 'bar')
```

# 4. Which player has won most number of medals? (Top 5)#

# 7. Which event is most popular in terms on number of players? (Top 5)

```python
tp = df
del tp['Year']
del tp['City']
del tp['Discipline']
del tp['Sport']
del tp['Country']
del tp['Gender']
del tp['Medal']
```

```python
x = list(df['Event'].unique())
l1 = df.values.tolist()
x = list(set(x))
l2=[]
for i in l1:
    l2.append(i[1])

most_event=[]
for i in x:
    most_event.append([i,l2.count(i)])

most_event = pd.DataFrame(most_event,columns = ['Event','No of Players'])
most_event
```

# 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df=pd.read_csv("summer.csv")
```

```python
female=df[df['Gender']=='Women']
female=female[female['Medal']=='Gold']
```

```python
fgm = female['Sport'].values.tolist()
fgm
fgm1 = list(set(fgm))

cnt = []
for i in fgm1:
    cnt.append([i,fgm.count(i)])

cnt = pd.DataFrame(cnt, columns =['Sport', 'No of Female Gold Medalists'])
cnt
```
